### 3. Describe three different data forms.

##### ANSWER:

Integers : Integers is represented by int(). It contains positive or negative whole numbers (without fraction or decimal). In Python there is no limit to how long an integer value can be.

### 10. What is the error caused by this expression? What would you do about it?

'I have eaten ' + 99 + ' burritos.'


##### ANSWER:

```python
'I have eaten ' + 99 + ' burritos.'
#the above mentioned expression causes an error because 99 is an integer,and only strings can be concatenated to other strings with the + operator.
```

```python
#the error in the above mentioned expression can be solved by converting 99 which is an integer to a string by using the string function as str(99):
'I have eaten ' + str(99) + ' burritos.'
```
